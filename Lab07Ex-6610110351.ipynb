```python
#Ex.1 โปรแกรมค้นหาคำในสตริง s1 ที่มี suffix หรือตัวอักษรลงท้าย โดยมีการรับค่า suffix จากอินพุต
s1 = 'apple one stable banana application station hyena two stay'
s2 =s1.split()
suffix = input('Suffix = ')
lst =[]
for x in s2 :
  if x.endswith(suffix) :
    lst.append(x)
for x in lst :
  print(x , end=' ')
```

```python
#Ex.3 โปรแกรมรับค่าสตริง s1 แทนการกำหนดค่าเริ่มต้นในโปรแกรม และรับค่า substr ที่ต้องการค้นหาแล้วแสดงผลลัพธ์เป็นรายการของคำที่เจอ substring นั้น
s1 = input().split()
substr = input()
lst = []
for x in s1 :
  if substr in x :
    lst.append(x)
for x in lst :
  print(x , end=' ')
```
