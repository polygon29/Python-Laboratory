```python
#Ex.1 จงเขียนโปรแกรมรับค่า n และแสดงสูตรคูณ ของ n จาก 1 ถึง 10
n = int(input('n = '))
for e in range(1,11):
  print(F'{n} x {e}')
```

```python
#Ex.2 จงเขียนโปรแกรมรับค่า n ที่มากกว่า 1 แล้วสร้างลิสต์ของอนุกรม 1^2,2^2,3^2,...,n^2 และหาผลรวม
n = int(input('n = '))
lst = []
for x in range(1,n+1) :
   x = x*x
   lst.append(x)
print(lst,sum(lst))
```

```python
#Ex.3 จงเขียนโปรแกรมรับค่า n ที่มากกว่า 1 แล้วสร้างลิสต์ของอนุกรมเลขคี่ จาก 1 ถึง n และหาผลรวม
n = int(input('n = '))
lst = []
for x in range(1,n+1):
  if x%2 == 1:
    lst.append(x)
print(lst,sum(lst))
```

```python
#Ex.4 จงเขียนโปรแกรมรับค่าสตริง แล้วให้นับตัวอักษรชนิดตัวพิมพ์เล็กและตัวพิมพ์ใหญ่เพิ่มด้วย
string = input('Enter a string = ')
#Number of digits
digit = '0123456789'
d = 0
for x in string:
  if x in digit:
    d += 1
print(F'Number of digits = {d}')
#Number of small letters
small = 'abcdefghijklmnopqrstuvwxyz'
m = 0
for y in string:
  if y in small:
    m += 1
print(F'Number of small letters = {m}')
#Number of capital letters
Big = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
n = 0
for z in string:
  if z in Big:
    n += 1
print(F'Number of capital letters = {n}')
```

```python
#Ex.5 จงเขียนโปรแกรมรับค่าคะแนนเข้ามา จนกว่าจะมีการใส่ค่าคะแนนเป็นศูนย์หรือค่าติดลบจึงหยุด แล้วให้หาค่าเฉลี่ยของคะแนนทั้งหมด
print('Enter scores (exit with score = 0)')
score = []
s = 1
while s != 0:
  s = input('score = ')
  s = float(s)
  if s>0:
    score.append(s)
  else :
    break
print(score)
print(F'Average score = {sum(score)/len(score):.2f}')
```
