```python
#Ex.1 a) A program to read only partial parts from the file.
f = open("sample.txt", "r")
print(f.read(6))
f.close()
```

```python
#Ex.1 b) A program to read an entire text file.
f = open("sample.txt", "r")
print(f.read())
f.close()
```

```python
#Ex.1 c) A program to read the number of lines from a text file.
f = open("sample.txt", "r")
print(f.readline())
print(f.readline())
f.close()
```

```python
#Ex.2 เติมคำสั่งในโปรแกรมเพื่อแสดงตัวอักษร 6 ตัวจากไฟล์ที่ต้องการ
from pathlib import Path
path = Path("sample.txt")
f = path.open(mode = "r")
print(f.read(6))                  # Read 6 characters from file
f.close()                            # close file
```

```python
#Ex.2 a) Using with statement and try-except statement to write a file
from pathlib import Path
try :
  fpath = Path("sample.txt")
  with fpath.open(mode = "r") as f :
    print(f.read())
except :
  print('Something went wrong.')
```

```python
#Ex.3 Reading an entire text file and displaying them nicely on the screen (ใช้ฟังก์ชัน strip() ตัดค่า \n ออกจากสมาชิกของลิสต์)
from pathlib import Path

fpath = Path("sample.txt")
try:
  with fpath.open(mode = "r") as f :
    lst = f.readlines()
    for line in lst :
      print(line.strip())
except :
  print("Something went wrong.")
```

```python
#Ex.4 Writing names to a file
from pathlib import Path

names = ['Babara\n', 'Annabel\n', 'Cathy\n', 'Dorothy\n']

fpath = Path("myfriends.txt")
try :
  with fpath.open(mode = "w") as f :
    f.writelines(names)
except :
  print('Something went wrong.')
```
