```python
#A3. Count Pass/Fail จากโปรแกรม Ex.5 เมื่อคำนวณค่าเฉลี่ยคะแนนเรียบร้อยแล้ว ให้ตรวจสอบค่าคะแนนของทุกคนว่ามีจำนวนคนผ่านกี่คน และคนตกกี่คน โดย Pass คือคนที่ได้คะแนนไม่ต่ำกว่าค่าเฉลี่ย และ Fail คือคนที่ได้คะแนนต่ำกว่าค่าเฉี่ย
print('Enter scores (exit with score = 0)')
score = []
s = 1
while s != 0:
  s = input('score = ')
  s = float(s)
  if s>0:
    score.append(s)
  else :
    break
averagescore = sum(score)/len(score)
print(score)
print(F'Average score = {averagescore:.2f}')
Pass = 0
Fail = 0
for x in score:
  if x<averagescore:
    Fail += 1
  else :
    Pass += 1
print(F'Pass = {Pass}, Fail = {Fail}')
```
