```python
#Ex.1
#สร้างอาเรย์ M ขนาด 2 มิติ ขนาด 8x8 ขึ้นมาจาก x อาเรย์ 1 มิติที่มีสมาชิก 64 ตัว
#จงเพิ่มโค้ดให้รับอินพุตเป็นจำนวนเต็มบวก 1 จำนวนจากแป้นพิมพ์แล้วเก็บใน k
#หลังจากนั้นให้เปลี่ยนค่าของช่องใน M ที่หมายเลขแถวและหลักหารด้วย k ลงตัว ให้มีค่าเป็น 0 แล้วสั่งพิมพ์ค่าอาเรย์ M ออกมา

import numpy as np
x = np.arange(64)
print('x = ',x)
M = np.array(x)
M = M.reshape(8,8)
print('M = ',M)

k = int(input('k = '))
M[::k , ::k] = 0
print(M)
```

```python
#Ex.2
#จงสร้าง P อาเรย์ 2 มิติ ชนิด int ขนาด 10x10 โดยให้สมาชิกทุกตัวมีค่าเป็น 1 โดยใช้ฟังก์ชัน
#แล้วเขียนโค้ดรับอินพุตเป็นจำนวนเต็มบวก 1 จำนวนจากแป้นพิมพ์ แล้วเก็บใน k
#หลังจากนั้นให้เปลี่ยนค่าของช่องใน P ที่หมายเลขแถวและหลักหารด้วย k ลงตัว ให้มีค่าเป็น k เท่าของค่าเดิม โดยป้อนค่า k อยู่ในช่วง 1-9
import numpy as np
P = np.ones(100, int)
P = np.array(P)
P = P.reshape(10,10)
print('P = ',P)
k = int(input('k = '))
P[::k , ::k] = k
print(P)
```

```python
#Ex.3
#สร้าง M เป็นอาเรย์ 2 มิติ ชนิด float ขนาด 5x5 ด้วยฟังก์ชัน identity() และปรับค่าสมาชิกดังในโค้ด
#จงเขียนโค้ดหาค่าผลต่างของค่ามากที่สุดและค่าน้อยที่สุดในแต่ละหลัก

import numpy as np
M = np.identity(5 , float)
print(M)
M[::2 , ::3] = 8
print(M)
M[2:4 , 1:4] = [5,10,15]
print(M)
N1 = np.max(M , axis = 0)
N2 = np.min(M , axis = 0)
N = N1-N2
print(N)
```

```python
#Ex.4
#อาเรย์ X เก็บความยาวของด้านประกอบมุมฉากของรูปสามเหลี่ยมมุมฉากหลายรูป
#ให้สร้างอาเรย์ Y ซึ่งเก็บความยาวด้านตรงข้ามมุมฉาก
#แสดงผลค่าสมาชิกในอาเรย์ X และ Y
#จากนั้นให้รวมอาเรย์ X และ Y เข้าด้วยกันในอาเรย์ Z ซึ่งเก็บความยาวด้านทั้ง
#แสดงผลค่าสมาชิกในอาเรย์ Z
#แล้วเขียนอาเรย์ลงในไฟล์ไบนารี .pny ชื่อ 'ztriangles' และอ่านเก็บขึ้นมาในอาเรย์ W
#เขียนอาเรย์ W ลงในไฟล์ตัวอักษร .txt ชื่อ 'wtriangles'

import numpy as np
X = np.array([[3,4],[5,12],[24,7],[1,1]])
print(X)

Y = []
Z = []
for i in X :
  c = (i[0]**2+i[1]**2)**0.5
  Y.append(c)
  Z.append([i[0] , i[1] , c])
Y = np.array(Y)
Z = np.array(Z)
print(Y)
print(Z)

np.save('ztriangles' , np.array(Z))
W = np.load('ztriangles.npy')
np.savetxt('wtriangles.txt' , W , delimiter=' ')
```
