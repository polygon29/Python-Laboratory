```python
#A3 Sum of Digits โปรแกรมรับค่าจำนวนเต็มบวก (n) หนึ่งค่า ซึ่งมีค่าไม่เกิน 10 หลัก แล้วให้คำนวณหาผลรวมของตัวเลขทุกหลัก
n = int(input('Enter n: '))
n1 = n%1e9
n2 = n1%1e8
n3 = n2%1e7
n4 = n3%1e6
n5 = n4%1e5
n6 = n5%1e4
n7 = n6%1e3
n8 = n7%1e2
n9 = n8%1e1
N0 = n//1e9
N1 = n1//1e8
N2 = n2//1e7
N3 = n3//1e6
N4 = n4//1e5
N5 = n5//1e4
N6 = n6//1e3
N7 = n7//1e2
N8 = n8//1e1
N9 = n9
Ntotal = N0+N1+N2+N3+N4+N5+N6+N7+N8+N9
print('Sum of digits =' , Ntotal)
```
