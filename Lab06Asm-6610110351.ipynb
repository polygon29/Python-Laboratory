```python
#A3 โปรแกรมรับค่า m และแสดงตัวเลขอนุกรม Fibonacci : a_0,a_1,a_2,...,a_n โดยมีค่าไม่เกิน m กำหนดให้ a_0=0 และ a_1=1 และ a_i=a_(i-2)+a_(i-1)
m = int(input('m = '))
a = [0,1]
i = 2
x = a[i-2]+a[i-1]
while x <= m :
  a.append(x)
  i += 1
  x = a[i-2]+a[i-1]
n = len(a)
for i in range(n) :
  print(a[i] , end=' ')
  if i<n-1 :
    print(',' , end=' ')
  else :
    print()
```
