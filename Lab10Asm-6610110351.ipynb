```python
#A3. Copying contents from one file to the other, except line 3 and 5
from pathlib import Path

imfile = input('Enter a data file to copy from: ')
fimfile = Path(imfile)

fpath = Path("a3in.txt")

import os
try :
    with fimfile.open(mode = "r") as f :
      a = f.readlines()
      outfile = input('Enter a data file to write from: ')
      foutfile = Path(outfile)
      print(F'The file ({imfile}) has {len(a)} lines.')
      for i in range(len(a)) :
        if i == 2 or i == 4 :
          pass
        else :
          if i == 0 :
            with foutfile.open(mode = "w") as g :
              b = a[i].strip('\n')
              g.writelines(b)
          else :
            with foutfile.open(mode = "a") as g :
              b = a[i].strip('\n')
              g.writelines('\n' + b)
      print(F'Now, they are copied to a file ({outfile})')
except :
  print(F'The input file ({imfile}) does not exist!')
```
