```python
#Ex.1
#จงเขียนโปรแกรม เลือกข้อมูลของประเทศใดประเทศหนึ่ง และสรุปจำนวนเหรียญรางวัลโอลิมปิกจากตารางข้อมูล ให้สร้างเป็นกราฟแบบเส้นและจุด

#Japan

#import
import matplotlib.pyplot as plt
import numpy as np
#==================

#data
gold = np.array([7,12,27])
silver = np.array([14,8,14])
bronze = np.array([17,21,17])
total = np.array([38,41,58])
year = np.array(["2012" , "2016" , "2020"])
#==========================

plt.title('Olympic Medals : Japan')
plt.xlabel('Olympic Games Year')
plt.ylabel('Medal Counts')
plt.plot(year,gold, marker='o', color='r', label='Gold')
plt.plot(year,silver, marker='o', color='b', label='Silver')
plt.plot(year,bronze, marker='o', color='g', label='Bronze')
plt.plot(year,total, marker='o', color='k', label='Total')
plt.grid()
plt.legend()
plt.show()
```

```python
#Ex.2
#จงเขียนโปรแกรม เลือกข้อมูลของ 2 ประเทศ และสรุปจำนวนเหรียญรางวัลโอลิมปิกจากตารางข้อมูล ให้สร้างเป็นกราฟแบบจุด แยกเป็นกราฟย่อย 4 กราฟ ตามประเภทของเหรียญ (Gold, Silver, Bronze และ Total)

#import
import matplotlib.pyplot as plt
import numpy as np
#==================

#data
#Japan
Jgold = np.array([7,12,27,0])
Jsilver = np.array([14,8,14,0])
Jbronze = np.array([17,21,17,0])
Jtotal = np.array([38,41,58,0])
#USA
Ugold = np.array([46,46,39,0])
Usilver = np.array([28,37,41,0])
Ubronze = np.array([30,38,33,0])
Utotal = np.array([104,121,113,0])
year = np.array(["2012" , "2016" , "2020" , "next"])
#=================================

#subplot 1 Total
plt.subplot(2,2,1)
plt.title('Olympic Medals : USA vs. Japan')
plt.ylabel('Medal Counts')
plt.plot(year, Utotal, '*b', label='USA Total')
plt.plot(year, Jtotal, 'or', label='Japan Total')
plt.grid()
plt.legend(loc='lower right')

#subplot 2 Gold
plt.subplot(2,2,2)
plt.title('Olympic Medals : USA vs. Japan')
plt.plot(year, Ugold, '*b', label='USA Gold')
plt.plot(year, Jgold, 'or', label='Japan Gold')
plt.grid()
plt.legend(loc='lower right')

#subplot 3 Silver
plt.subplot(2,2,3)
plt.xlabel('Olympic Games Year')
plt.ylabel('Medal Counts')
plt.plot(year, Usilver, '*b', label='USA Silver')
plt.plot(year, Jsilver, 'or', label='Japan Silver')
plt.grid()
plt.legend(loc='lower right')

#subplot 4 Bronze
plt.subplot(2,2,4)
plt.xlabel('Olympic Games Year')
plt.plot(year, Ubronze, '*b', label='USA Bronze')
plt.plot(year, Jbronze, 'or', label='Japan Bronze')
plt.grid()
plt.legend(loc='lower right')

plt.show()
```

```python
#Ex.3
#จากตารางข้อมูลเหรียญรางวัลโอลิมปิก ให้เลือกข้อมูลเพื่อนำเสนอเป็นแผนภูมิแท่ง (bar graph) อย่างเหมาะสม

#import
import matplotlib.pyplot as plt
import numpy as np
#==================

#data
country = np.array(['USA' , 'China' , 'Japan' , 'Great Britain' , 'Russia' , 'Australia' , 'Germany'])
total = np.array([113,88,58,65,71,46,37])
#==========================================================

plt.title('Olympic Games 2020 : Total Medals')
plt.bar(country, total, color='g')
plt.show()
```

```python
#Ex.4
#จากตารางข้อมูลเหรียญรางวัลโอลิมปิก ให้เลือกข้อมูลเพื่อนำเสนอเป็นแผนภูมิวงกลม (pie chart) อย่างเหมาะสม

#import
import matplotlib.pyplot as plt
import numpy as np
#==================

#data
country = np.array(['USA' , 'China' , 'Japan' , 'Great Britain' , 'Russia' , 'Australia' , 'Germany'])
total = np.array([113,88,58,65,71,46,37])
#==========================================================

plt.title('Olympic Games 2020 : Total Medals')
plt.pie(total, labels=country)
plt.show()
```
