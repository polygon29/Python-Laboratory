```python
#A1. Shopping Cart in a book shop

# Class Book
class Book:
    def __init__(self, title, price, keywords):
        self.title=title; self.price=price; self.keywords=set(keywords)
    def __str__(self):
        return self.title + ' ('+ str(self.price)+' THB)'
    def __lt__(self, bk):
        if self.price != bk.price:
            return self.price < bk.price
        else:
            return self.title < bk.title
    def update_price(self, new_price):
        self.price = new_price

class EBook(Book):
  # constructor:
  def __init__(self, title, price, keywords, ebformat):
    self.title=title
    self.price=price
    self.keywords=set(keywords)
    self.ebformat = ebformat
  def __str__(self):
        st = self.title
        st += ' [e-book/'+ self.ebformat +']'
        st += ' ('+ str(self.price)+' THB)'
        return st

# Class ShoppingCart
class ShoppingCart:
    def __init__(self, id):
        self.id = id
        self.books = [] # list of books in the shopping cart
        # Each item is [book, number_of_books]
        # for example [[b4,1],[b1, 2],[b2, 1]]

    def add_book(self, book):  # add only 1 book at a time
        found = False
        idx = 0;
        for b in self.books: # find the book in the cart
            if b[0]==book :
                found = True
                x = idx  # position to be added
                break
            idx += 1
        if found : # the same book found in the cart
            bk = self.books[x][0]
            n = self.books[x][1]
            ab = [bk, n+1]  # increase number of the same book
            self.books[x]= ab # update the book item
        else : # add a new book [bk, 1] into the list
            bk = book
            self.books.append([bk,1]) # add a new item

    def get_total(self):
        t = 0.0
        for b in self.books :
          t += b[0].price*b[1]
        return t

    def __str__(self):
        strcart = 'Cart ID: '+ str(self.id)+'\n'
        p = 0
        if len(self.books)==0:
            strcart += 'The cart is empty\n'
        else:
            for b in self.books:
                strcart += str(b[0])+' x '+ str(b[1])+'\n'
                p += b[1]*b[0].price
        strcart += F'Total price = {p} THB'
        strcart += '\n----------------------\n'
        return strcart

# ------- main program --------
b1 = Book('Python-101', 200, ['python','programming','coding'])
b2 = Book('Calculus 1', 250, ['math','calculus'])
b3 = Book('Calculus 2', 300, ['math','calculus'])
b4 = Book('Physics 1', 280, ['physics','science'])
b5 = Book('Chemistry 1', 200, ['chemistry','science'])
eb1 = EBook('Python 101', 150, ['python','programming','coding'],'PDF')
eb2 = EBook('Advanced Calculus', 400, ['math','calculus'],'EPUB')

allbooks = [b1,b2,b3,b4,b5,eb1,eb2]
allbooks.sort()
print('List of all books with Book_id')
n = len(allbooks)
for bkid in range(n):
    print(F"Book_id = {bkid}: {allbooks[bkid]}")

mycart = ShoppingCart(650903012)
print('This is a new shopping cart')
print(mycart)
print('Add a book into the cart by Book_id')
print('Enter an invalid Book_id for stop adding')
bkid=1
while bkid>=0:
    bkid = int(input('Enter Book_id: '))
    if 0<= bkid < len(allbooks):
        bk = allbooks[bkid]
        mycart.add_book(bk)
        print(F'"{bk}" is added')
    else :
        print('Invalid Book_id ==> Done adding\n')
        break

print('Shopping Cart Status')
print(mycart)

print('Discount 10%')
total = mycart.get_total()
print(F'Cart ID: {mycart.id}\nTotal = {total} THB')
# Edit แก้ไข total = คำนวณราคาหักส่วนลดแล้ว
total = total*90/100
print(F'After discount : {total} THB')

```
