```python
#Ex.1 โปรแกรมรับค่า n และแสดงตัวเลขอนุกรม a_1,a_2,...,a_n กำหนดให้ a_i=2a_(i-1)+i โดยมีค่าเริ่มต้น a_1=1
n = int(input('n = '))
a = [0,1]
for i in range(2,n+1):
  x = 2*a[i-1]+i
  a.append(x)
for i in range(1,n+1):
  print(a[i], end=' ')
  if i<n:
    print(',', end=' ')
  else :
    print()
```

```python
#Ex.2 โปรแกรมรับค่า n และแสดงตัวเลขอนุกรม a_1,a_2,...,a_n กำหนดให้ a_i=a_(i-1)+2i-1 โดยมีค่าเริ่มต้น a_1=0
n = int(input('n = '))
a = [-1,0]
#เก็บอนุกรม
for i in range(2,n+1):
  x = a[i-1]+2*i-1
  a.append(x)
#printอนุกรม
for i in range(1,n+1):
  print(a[i] , end=' ')
  if i<n:
    print(',' , end=' ')
  else :
    print()
```

```python
#Ex.3 โปรแกรมรับค่า m และแสดงตัวเลขอนุกรม a_1,a_2,...,a_n โดยมีค่าไม่เกิน m กำหนดให้ a_i=2a_(i-1)+i โดยมีค่าเริ่มต้น a_1=1
m = int(input('m = '))
a = [0]
x = 1; i = 1
while x <= m:
  a.append(x)
  i += 1
  x = 2*a[i-1]+i
n = len(a)
for i in range(1,n):
  print(a[i] , end=' ')
  if i<n-1:
    print(',' , end=' ')
  else :
    print()
```

```python
#Ex.4 โปรแกรมรับค่าสตริงบรรทัดหนึ่ง แล้วนำมาเก็บในลิสต์ของสตริง (ลิสต์ของคำ)
word = input('Enter words: ').split()
print(word)
n = len(word)
print(F'There are {n} words')
print('Alphabetical order')
word.sort()
print(word)
```

```python
#Ex.5 โปรแกรมรับค่าจำนวนเต็มมาเก็บไว้ในลิสต์แล้วเรียงลำดับใหม่จากมากไปน้อย
nums = [int(x) for x in input('Enter integers: ').split()]
print(nums)
n = len(nums)
print('There are' , n , 'integers')
print('Descending order')
nums.sort(reverse=True)
print(nums)
odd = 0
even = 0
for i in range(n) :
  if nums[i]%2 == 0 :
    even += 1
  else :
    odd += 1
print('Even numbers =' , even)
print('Odd numbers =' , odd)
```
