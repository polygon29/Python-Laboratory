```python
#(1)  Ex.1 โปรแกรมรับความกว้างและความยาวของรูปสี่เหลี่ยมผืนผ้า (Rectangle) แล้วคำนวณหาพื้นที่ (Area) และความยาวเส้นรอบรูป (Perimeter)
a,b = input('Rectangle width and length: ').split()
a = float(a)
b = float(b)
area = a*b
perimeter = 2*a+2*b
print('Area =' , area , ',' , 'Perimeter =' , perimeter)
```

```python
#(2)  Ex.2 โปรแกรมแปลงอุณหภูมิระหว่าง 3 หน่วย จากองศาเซลเซียส เป็นองศาฟาเรนไฮต์และเคลวิน
a = input('Enter temperature in Celsius: ')
a = float(a)
fah = a*9/5+32
kel = a+273.5
print(a , 'Celsius =' , fah , 'Fahrenheit =' , kel , 'Kelvin')
```

```python
#(3)  Ex.3 โปรแกรมแปลงหน่วยพื้นที่ เฮกตาร์ (Hectare) เป็น ตารางกิโลเมตร (Square km) และ เอเคอร์ (Acres)
a = input('Enter area in  Hectare: ')
a = float(a)
sq_km = a*0.01
acr = a*2.471
print(a , 'hectare =' , sq_km , 'sq km =' , acr , 'acres')
```

```python
#(4)  Ex.4 โปรแกรมรับค่าความยาวด้าน 3 ด้านของรูปสามเหลี่ยม ให้ใส่ด้านที่ยาวที่สุดเป็นค่าสุดท้าย แล้วโปรแกรมตรวจสอบว่าเป็นด้านของรูปสามเหลี่ยมมุมฉากหรือไม่ ให้พิมพ์ออกมาว่า True หรือ False
a,b,c = input('Enter 3 sides of a triangle, the last one must be the longest side: ').split()
a = float(a)
b = float(b)
c = float(c)
x = c*c==a*a+b*b
print(x)
```

```python
#(5)  Ex.6 โปรแกรมสอบถามชื่อและปีเกิดของบุคคล จากนั้นใส่ค่าปีปัจจุบันแล้วให้แสดงอายุของบุคคลนั้น
name = input('Enter name: ')
birth = input('Enter birth year: ')
current = input('Enter current year: ')
birth = int(birth)
current = int(current)
year = current-birth
print('This year,' , name , 'is' , year , 'years old')
```
