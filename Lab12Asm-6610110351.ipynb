```python
#A2.
#เขียนโปรแกรมอ่านไฟล์ worldpop1951_2020.txt
#นำเสนอเป็นกราฟแสดงจำนวนประชากรในทุกๆ 5 ปี

#import
import matplotlib.pyplot as plt
import numpy as np
#=====================

fname = 'worldpop1951_2020.txt'
with open(fname) as f :
  s = f.read()

ls = s.split()
lsint = [int(e) for e in ls]
n = len(lsint)
Yr = [] ; Pop = []
for i in range(0,n,2) :
  Yr.append(lsint[i])
  Pop.append(lsint[i+1])
Year = np.array(Yr)
P = np.array(Pop)
Y10 = Year[4::5]
PM = P[4::5]
IncrPop = 0.000001*PM

plt.title('World Population in every 5 years')
plt.ylabel('Population (Millions)')
plt.bar(Y10,IncrPop , color='g')
plt.plot(Y10,IncrPop , marker='*' , color='b')
plt.grid()
plt.show()
```
