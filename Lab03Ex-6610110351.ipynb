```python
#(2) Ex.2 โปรแกรมรับข้อความหนึ่งบรรทัด เก็บเป็นสตริงตัวเดียว แล้วให้แสดงค่าความยาวของข้อความนั้น (string length) ตามด้วย ค่าตำแหน่งกึ่งกลางและพิมพ์ตัวอักษรที่อยู่ตำแหน่งกึ่งกลางของสตริงโดยให้ใส่เครื่องหมาย ' ' คร่อมตัวอักษรนั้นด้วย
s = str(input('Enter your string: '))
s = list(s)
s1 = (len(s)+1)//2
s0 = s1-1
st = (s[s0])
print(len(s) , s1 , "'" + st + "'")
```

```python
#(3) Ex.3 โปรแกรมรับข้อความหนึ่งบรรทัด เก็บเป็นสตริงตัวเดียวซึ่งเป็นรหัส ชื่อและนามสกุลของนักศึกษา แล้วให้พิมพ์ค่าออกมา 3 บรรทัดเป็นข้อความนั้นแบบเดิม ข้อความนั้นแบบเป็นตัวอักษรพิมพ์เล็กทั้งหมด (lower) และข้อความนั้นแบบตัวอักษรพิมพ์ใหญ่ทั้งหมด (upper)
s = str(input('Enter your information: '))
s1 = s.lower()
s2 = s.upper()
print(s)
print(s1)
print(s2)
```

```python
#(4) Ex.4 โปรแกรมรับข้อความหนึ่งบรรทัดเก็บเป็นลิสต์ของคำที่ไม่รวมช่องว่างแล้วให้แสดงจำนวนคำที่นับได้ แล้วตามด้วยคำแรกและคำสุดท้ายที่ใส่เข้ามาโดยให้ใส่เครื่องหมาย ' ' คร่อมคำด้วย
s = input('Enter words:' ).split()
print(len(s) , "'" + s[0] + "'" , "'" + s[-1] + "'")
```

```python
#(5) Ex.5 โปรแกรมรับค่าจำนวนเต็มบวก n แล้วให้สร้างลิสต์ของจำนวนเต็ม จาก 1 ถึง n แล้วให้โปรแกรมพิมพ์ลิสต์นั้นขึ้นมา พร้อมทั้งแสดงค่าผลรวมจาก 1 ถึง n
s = int(input('n = '))
s1 = range(1, s+1)
s1 = list(s1)
print(s1)
print('Summation = ' , sum(s1))
```
