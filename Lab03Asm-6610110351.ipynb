```python
#A1. Score จงเขียนโปรแกรมรับค่าคะแนน(จำนวนจริง)ของการสอบครั้งหนึ่ง มีคะแนนของนักเรียนจำนวนหนึ่ง ให้โปรแกรมรับอินพุตข้อมูลคะแนนของทุกคนในหนึ่งบรรทัด คั่นด้วยช่องว่าง ให้เก็บไว้เป็นลิสต์ของ float จากนั้นให้พิมพ์รายงานผลจำนวนนักเรียนทั้งหมด คะแนนต่ำสุด คะแนนสูงสุด และค่าเฉลี่ย
s = [float(x) for x in input('Input all scores: ').split()]
#s เป็นลิสต์
av = sum(s)/len(s)
print('Number of students =' , len(s))
print('Min score =' , min(s))
print('Max score =' , max(s))
print(F'Average = {av:.2f}')
```
