```python
#Ex.1แก้ไขเพิ่มเติมโปรแกรมนี้ เพิ่มเติมส่วนของนิยามฟังก์ชันให้สมบูรณ์และทำงานได้ตามคำบรรยาย

#นิยามฟังก์ชัน คำนวณพื้นที่รูปสี่เหลี่ยมผืนผ้า
def rectangleArea(W, L) :
  Area = W*L
  return Area

#นิยามฟังก์ชัน คำนวณเส้นรอบรูปสี่เหลี่ยมผืนผ้า
def rectanglePerim(W, L) :
  Perim = W+W+L+L
  return Perim

#ส่วนหลักของโปรแกรม
print('Rectangle')
wd = float(input('Width = '))
ln = float(input('Length = '))
a = rectangleArea(wd, ln)
print(F'Area rectangle ({wd}x{ln})')
print(F'Area = {a}')
p = rectanglePerim(wd, ln)
print(F'Perimeter = {p}')
```

```python
#Ex.2 โปรแกรมคำนวณพื้นที่รูปสามเหลี่ยมความยาวด้าน a,b,c จงเพิ่มเติมส่วนหลักของโปรแกรมให้มีอินพุตข้อมูลและเรียกฟังก์ชันให้ถูกต้อง
import math

#นิยามฟังก์ชัน คำนวณพื้นที่รูปสามเหลี่ยมจากความยาวด้าน a,b และ c
def TriangleArea(a, b, c) :
  s = (a+b+c)/2
  s1 = s*(s-a)*(s-b)*(s-c)
  if (s1>0) :
    area = math.sqrt(s1)
  else :
    area = 0
  return area

#ส่วนหลักของโปรแกรม
print('Triangle')
a = float(input('a = '))
b = float(input('b = '))
c = float(input('c = '))
ar = TriangleArea(a, b, c)
print(F'Area of Triangle ({a},{b},{c}) = {ar:.4f}')
```

```python
#Ex.3 แก้ไขเพิ่มเติมโปรแกรมนี้ โดยเพิ่มเติมส่วนของนิยามฟังก์ชัน printAll ให้สมบูรณ์

#นิยามฟังก์ชัน printAll
def printAll(s) :
  x = s.upper()
  y = s.lower()
  print(x)
  print(y)

#Main program
a = input('Enter string: ')
printAll(a)
```
