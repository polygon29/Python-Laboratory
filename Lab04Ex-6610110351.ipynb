```python
#Ex.1  รับจำนวนเต็มเข้ามาสองตัว แล้วตรวจสอบค่าตัวแรกมากกว่าตัวที่สอง หรือน้อยกว่า หรือว่าค่าเท่ากัน
a,b = input('Enter 2 numbers: ').split()
a = int(a)
b = int(b)
if a>b:
  print(F'{a} is greater than {b}')
elif a==b:
  print(F'{a} is equal to {b}')
else :
  print(F'{a} is less than {b}')
```

```python
#Ex.2 โปรแกรมรับค่าจำนวนเต็ม 3 ค่าแล้วแสดงค่าที่มากที่สุด
a,b,c = [int(x) for x in input('Enter 3 numbers: ').split()]
max = a
if max<b:
  print(F'Max = {b}')
elif max<c:
  print(F'Max = {c}')
else :
  print(F'Max = {a}')
```

```python
#Ex.3 รับจำนวนเต็ม 1 ค่า แล้วตรวจสอบว่าเป็นค่าศูนย์ หรือเลขคู่บวก หรือเลขคู่ลบ หรือเลขคี่บวก หรือเลขคี่ลบ
n = int(input('Enter an integer: '))
if n==0:
  print(F'{n} is zero')
else :
  if n>0:
    print(F'{n} is a positive' , end=' ')
  else :
    print(F'{n} is a negative' , end=' ')
  if n%2==0:
    print('even' , end=' ')
  else :
    print('odd' , end=' ')
  print('number')
```

```python
#Ex.4 โปรแกรมรับค่าจำนวนจริงบวก 3 ค่า แล้วตรวจสอบว่าเป็นด้านของรูปสามเหลี่ยมหรือไม่ โปรแกรมพิมพ์ A Triangle ถ้าเป็นรูปสามเหลี่ยม หรือ Not A Triangle ถ้าไม่ใช่รูปสามเหลี่ยม
a,b,c = input('Enter 3 numbers: ').split()
a = float(a)
b = float(b)
c = float(c)
if a>b+c:
  print('Not A Triangle')
elif b>c+a:
  print('Not A Triangle')
elif c>a+b:
  print('Not A Triangle')
else :
  print('A Triangle')
```

```python
#Ex.5 รับชื่อและนามสกุลของบุคคลหนึ่ง แล้วตรวจสอบว่า ชื่อยาวหรือสั้น โดยดูจากจำนวนตัวอักษรของชื่อและนามสกุล หากมีชื่อหรือนามสกุลที่ยาวกว่าหรือเท่ากับ 10 ตัวอักษร หรือชื่อและนามสกุลรวมกันมีตัวอักษรมากกว่า 15 ตัวอักษร ให้ถือว่าเป็นชื่อยาว ไม่เช่นนั้นก็เป็นชื่อสั้น
nam = input('Enter your name and last name: ')
name = nam.split()
t1 = list(name[0])
t2 = list(name[1])
if len(t1)>=10:
  print(F"'{nam}' is a long name")
elif len(t2)>=10:
  print(F"'{nam}' is a long name")
elif len(t1)+len(t2)>=15:
  print(F"'{nam}' is a long name")
else :
  print(F"'{nam}' is a short name")
```
