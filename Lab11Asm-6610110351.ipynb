```python
#A2.
#จงเขียนโปรแกรมอ่านความสูง (เซนติเมตร) และน้ำหนัก (กิโลกรัม) เป็นจำนวนจริง (float) มาเพื่อสร้าง numpy array แบบ 2 มิติ
#จงหาค่าอาเรย์ 1 มิติ เก็บ bmi
#หาค่าเฉลี่ย bmi ทั้งหมดที่คำนวณได้ เก็บใส่ตัวแปร avg_bmi
#นับจำนวน bmi ที่คำนวณได้ที่มีค่าน้อยกว่า 18.5

import numpy as np
list_hw = []
n  = int(input())
for k in range(n) :
  h,w = input().split()
  list_hw.append((float(h) , float(w)))
list_hw
hw = np.array(list_hw)
print('Height in cm and Weight in kg:\n' , hw)
bmi = (hw[:,1])/(hw[:,0]*0.01)**2
print('BMI:\n' , bmi)
avg_bmi = np.mean(bmi)
count_underweight = np.sum(bmi < 18.5)
print('average bmi =' , avg_bmi)
print('bmi < 18.5 =' , count_underweight)
```
