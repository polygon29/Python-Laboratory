```python
#A5. Triangle Area โปรแกรมรับจำนวนเต็มบวก 3 ค่า แล้วตรวจสอบว่าเป็นด้านของรูปสามเหลี่ยมหรือไม่ ถ้าไม่ใช่รูปสามเหลี่ยมให้พิมพ์ว่า Not A Triangle แต่ถ้าเป็นรูปสามเหลี่ยมให้คำนวณพื้นที่รูปสามเหลี่ยม แล้วพิมพ์ดังนี้ Triangle Area = ค่าพื้นที่รูปสามเหลี่ยมที่คำนวณได้ (แสดงค่าเป็นทศนิยม 1 ตำแหน่งเท่านั้น)
a,b,c = input('Enter 3 integers: ').split()
a = int(a)
b = int(b)
c = int(c)
#เซ็ทสูตร
s = (a+b+c)/2
area = (s*(s-a)*(s-b)*(s-c))**0.5
if a>b+c:
  print('Not A Triangle')
elif b>c+a:
  print('Not A Triangle')
elif c>a+b:
  print('Not A Triangle')
else :
  print(F'Triangle Area = {area:.1f}')
```
