```python
#Ex.1 เขียนเมธอดอัพเดตราคาหนังสือในคลาส Book ให้ถูกต้อง เพื่อให้ได้ผลการรันดังตัวอย่างนี้

# คลาส Book ข้อมูลตัวอย่างของหนังสือ ในร้านหนังสือ
class Book:
  # constructor:
  # สร้าง object Book ด้วยชื่อเรื่อง(title) ราคา(price) และ คำสำคัญ (keywords)
  def __init__(self, title, price, keywords):
        self.title=title
        self.price=price
        self.keywords=set(keywords) # set ของ keywords เก็บคำไม่ซ้ำ

  # method __str__ แปลง object เป็นสตริง(str) สำหรับพิมพ์ข้อมูลของ object
  def __str__(self):
        return self.title + ' ('+ str(self.price)+' THB)'
  # method __lt__  สำหรับการเปรียบเทียบ object มีการเรียกใช้เมื่อมีการ sort object นี้
  # เปรียบเทียบ แบบน้อยกว่า (less than) โดยเทียบจากราคาของหนังสือ
  # ถ้า ราคาเท่ากัน ให้เปรียบเทียบ title
  def __lt__(self, bk):
        if self.price != bk.price:
            return self.price < bk.price
        else:
            return self.title.lower() < bk.title.lower()
  # method สำหรับปรับเปลี่ยน ราคาของหนังสือ
  def update_price(self, new_price):
    self.price = new_price

# main program
b1 = Book('Python 101', 200, ['python','programming','coding'])
b2 = Book('Calculus 1', 200, ['math','calculus'])
b3 = Book('Calculus 2', 300, ['math','calculus'])

allbooks = [b1,b2,b3]
for b in allbooks:
    print(b)
print('-----After sorting books----')
allbooks.sort()
for b in allbooks:
    print(b)
dcpc = 5.0 # discount percentage ส่วนลดราคาคิดเป็นเปอร์เซ็นต์
for b in allbooks:
    p = b.price
    np = p*(100-dcpc)/100
    b.update_price(np)

print(F'-----After discount {dcpc}% ----')
for b in allbooks:
    print(b)
```

```python
#Ex.4 ให้เพิ่มการประกาศสร้างคลาส Ebook ซึ่งเป็น subclass ของ Book และเพิ่มข้อมูล e-book ให้ถูกต้อง

# คลาส Book ข้อมูลตัวอย่างของหนังสือ ในร้านหนังสือ
class Book:
  # constructor:
  # สร้าง object Book ด้วยชื่อเรื่อง(title) ราคา(price) และ คำสำคัญ (keywords)
  def __init__(self, title, price, keywords):
        self.title=title
        self.price=price
        self.keywords=set(keywords) # set ของ keywords เก็บคำไม่ซ้ำ

  # method __str__ แปลง object เป็นสตริง(str) สำหรับพิมพ์ข้อมูลของ object
  def __str__(self):
        return self.title + ' ('+ str(self.price)+' THB)'
  # method __lt__  สำหรับการเปรียบเทียบ object มีการเรียกใช้เมื่อมีการ sort object นี้
  # เปรียบเทียบ แบบน้อยกว่า (less than) โดยเทียบจากราคาของหนังสือ
  # ถ้า ราคาเท่ากัน ให้เปรียบเทียบ title
  def __lt__(self, bk):
        if self.price != bk.price:
            return self.price < bk.price
        else:
            return self.title.lower() < bk.title.lower()
  # method สำหรับปรับเปลี่ยน ราคาของหนังสือ
  def update_price(self, new_price):
        self.price = new_price

# ประกาศคลาส EBook  เป็น subclass ของคลาส Book
class EBook(Book) :
  # constructor:
  # สร้าง object Book ด้วยชื่อเรื่อง(title) ราคา(price) และ คำสำคัญ (keywords)
  # และ มี ebformat คือ e-book format เช่น 'PDF','EPUB','RTF'
  def __init__(self, title, price, keywords, ebformat):
    self.title=title
    self.price=price
    self.keywords=set(keywords)
    self.ebformat = ebformat
  # method __str__ แปลง object เป็นสตริง (str) สำหรับพิมพ์ข้อมูลของ object
  def __str__(self):
        st = self.title
        st += ' [e-book/'+ self.ebformat +']'
        st += ' ('+ str(self.price)+' THB)'
        return st

# main program
b1 = Book('Python 101', 200, ['python','programming','coding'])
b2 = Book('Calculus 1', 200, ['math','calculus'])
eb1 = EBook('Python 101', 150, ['python','programming','coding'],'PDF')
eb2 = EBook('Advanced Calculus', 400, ['math','calculus'],'EPUB')

allbooks = [b1,b2,eb1,eb2]
n = 0
for b in allbooks:
  n += 1
  print(F"{n}: {b}")
print('Enter a new e-book')
tt = input('Title: ')
p = float(input('Price: '))
kw = input('Keywords: ').split()
ebf = input('E-book format: ')
# nb = สร้างออบเจค EBook จากข้อมูลหนังสือใหม่
nb = EBook(tt,p,kw,ebf)
# เพิ่ม nb เข้าไปต่อท้ายลิสต์ allbooks
allbooks.append(nb)
# พิมพ์แจ้งข้อความ The new e-book "...." is added
print(F'The new book "{nb}" is added')

print('-----After sorting all books----')
allbooks.sort()
n = 0
for b in allbooks:
  n += 1
  print(F"{n}: {b}")
```
